#### Выделение клиентов по номерам телефонов, которые покупали не менее двух раз в течение квартала с предыдущей покупки за отчетный период

```python
def time_diff(df):
    # get the time differences and put them directly into the group
    df['time_since_previous'] = df.sort_values('order_date').order_date.diff()

    return df

df_reg = df.groupby('phone').apply(time_diff)

# df_reg =  df_reg.query('phone != 0', engine='python')
# df_reg
# df['time_since_previous'] = df.sort_values('order_date').groupby('phone').order_date.diff()
```

```python
df_reg.time_since_previous = df_reg.time_since_previous.dt.days

check_period = 90
df_reg.time_since_previous = df_reg.time_since_previous.fillna(0)
df_reg.time_since_previous = df_reg.time_since_previous.astype(int)
# df_reg = df_reg.query('0 <= time_since_previous < @check_period')
# df_reg
df_reg = df_reg.query('time_since_previous >0 ', engine='python')
df_reg.head()
```

```python
df_reg_interim = df_reg.groupby('phone', as_index=False).time_since_previous.mean()
df_reg_final = df_reg_interim.query('time_since_previous <= @check_period')
```

```python
df_reg_final
```
